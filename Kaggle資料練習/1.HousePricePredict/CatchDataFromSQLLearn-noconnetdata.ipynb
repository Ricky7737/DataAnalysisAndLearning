# 1.練習從SQL 抓取資料

```python
#練線SQL資料庫
Connet_SQL = pymysql.connect(host='放自己的資料',
                            port=3306,
                            user='root',
                            passwd='用自己的SQL密碼',
                            charset='utf8',
                            db='1-housepredictdata')
```

```python
#讀取SQL資料
#指定 table train
sql_train = 'select * from train'
data_train = pd.read_sql(sql_train, Connet_SQL)
```
