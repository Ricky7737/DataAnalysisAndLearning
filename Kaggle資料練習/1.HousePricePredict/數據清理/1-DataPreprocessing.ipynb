# 2.合併表格處理遺漏值

```python
#將資料合併，方便處理缺失值
#concat()默認為直向合併
#合併後Index要重置，透過 reset_index(drop=True) 來重置 
all_data = pd.concat((train_data, test_data)).reset_index(drop=True)
all_data.index = all_data["Id"]
all_data = all_data.drop("Id", axis= 1 )
all_data
```

# 4.補足遺漏值

```python
all_data.head(10)
```

```python
all_data.info()
```

```python
#PoolQC:游泳池等級,MiscFeature:其他類別未涵蓋的其他功能,Alley:胡同類型
#Fence:圍欄等級,FireplaceQu:壁爐等級
#語法補充 : fillna("要補入的數值") 用來填補遺漏值
for col in ("PoolQC","MiscFeature","Alley","Fence","FireplaceQu"):
    all_data[col] = all_data[col].fillna("None")
```

```python
#gropby 為分類指定欄位的資料
#lambda 類似 function 的語法
#以下就是將Neighborhood資料分類，同類資料裡的LotFrontage取中位數填補遺漏值
all_data["LotFrontage"] = all_data.groupby("Neighborhood")["LotFrontage"].transform(lambda x: x.fillna(x.median()))
```

```python
#接下處理有類似特徵的欄位，遺漏值以 None填補
#GarageCond:車庫條件,GarageQual:車庫等級
#GarageFinish:車庫內部裝潢,GarageType:車庫位置
for col in ("GarageCond","GarageQual","GarageFinish","GarageType"):
    all_data[col] = all_data[col].fillna("None")
```

```python
#GarageYrBlt:車庫建照年份,GarageArea:車庫面積,
#GarageCars:車庫可容納車子的數量
#接為數值型態 所以補0
for col in ("GarageYrBlt","GarageArea","GarageCars"):
    all_data[col] = all_data[col].fillna(0)
```

```python
#處理地下室 Bsmt 部分 字串資料補 None ; 數值型態補 0
#BsmtCond : 地下室的狀況,BsmtExposure:地下室採光程度,BsmtQual:地下室高度,
#BsmtFinType1:地下室1的等級,BsmtFinType2:地下室2的等
for col in ("BsmtCond","BsmtExposure","BsmtQual","BsmtFinType1","BsmtFinType2"):
    all_data[col] = all_data[col].fillna("None")
#BsmtFinSF1:地下室1的面積,BsmtFinSF2:地下室2的面積,
#BsmtHalfBath:地下室半浴室數量,BsmtFullBath:地下室全浴室數量,
#BsmtUnfSF:未完成的地下室面積,TotalBsmtSF:地下室總面積(為數值，所以缺值直接補0)
for col in ("BsmtFinSF1","BsmtFinSF2","BsmtHalfBath","BsmtFullBath","BsmtUnfSF","TotalBsmtSF",):
    all_data[col] = all_data[col].fillna(0)
```

```python
#接下來是「MSZoning」，中文意思為銷售的住宅類型
#以RL(低密度住宅)做替代，使用語法 .mode()[0]
#MasVnrType:砌體材質
all_data["MasVnrType"] = all_data["MasVnrType"].fillna("None")
#MasVnrArea:砌體面積
all_data["MasVnrArea"] = all_data["MasVnrArea"].fillna(0)
```

```python
#在合併後的數據裡，遺漏值只有2個，而唯一不同 NoSeWa 也只有一筆對於本資料分析無幫助
#因此刪除 "Utilties"
all_data = all_data.drop(["Utilities"], axis = 1)
```

```python
# Function 為家電等集，用「Typ」 典型來替代
all_data['Functional'] = all_data['Foundation'].fillna('Typ')
```

```python
# 而剩餘都只有一個遺漏值，因此都直接用最多的數值來填補
# pandas mode() 也就是取資料中頻率出現最多的 
all_data["Exterior1st"] = all_data["Exterior1st"].fillna(all_data["Exterior1st"].mode()[0])
all_data["Exterior2nd"]= all_data["Exterior2nd"].fillna(all_data["Exterior2nd"].mode()[0])
#SaleType:銷售類型，缺值一樣只有一個，所以用最普遍的WD(常規的契約)來替補
all_data["SaleType"] = all_data["SaleType"].fillna(all_data["SaleType"].mode()[0])
#Electrical:電力系統，缺值只有一個，所有用最普遍的SBrkr(標準的電力系統)來替補
all_data["Electrical"] = all_data["Electrical"].fillna(all_data["Electrical"].mode()[0])
#KitchenQual:廚房等級，缺值只有一個，所以用最普遍的TA(典型(中等)的等級)來替補
all_data["KitchenQual"] = all_data["KitchenQual"].fillna(all_data["KitchenQual"].mode()[0])
```

```python
#可以看到所有資料空值都背
all_data.isnull().sum().sort_values(ascending=False)
```
